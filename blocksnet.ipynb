индекс *Шенонна*

```python
from blocksnet import Diversity

diversity = Diversity(city_model=Villozskoe)
result_devirsity = diversity.calculate()
result_devirsity
```

```python
Diversity.plot(result_devirsity, figsize =(5,5))
```

Методы вычисления центральности кварталов города

```python
from blocksnet import Centrality, PopulationCentrality

centrality = Centrality(city_model=Villozskoe)
result_centrality = centrality.calculate()

result_centrality
```

```python
Centrality.plot(result_centrality, figsize =(5,5))
```

```python
centrality_population = PopulationCentrality(city_model=Villozskoe)
result_centrlity_population = centrality_population.calculate()
result_centrlity_population
```

```python
PopulationCentrality.plot(result_centrlity_population, figsize =(5,5))
```

Метрика оценки кварталов SpaceMatrix

```python
from blocksnet import Spacematrix

spacematrix = Spacematrix(city_model=Villozskoe)
result_spacematrix = spacematrix.calculate()

result_spacematrix
Spacematrix.plot(result_spacematrix, figsize =(12,12))
```

Метод определения типа землепользования квартала (LandUse) на основе сущетсвующих в модели городских сервисов.

```python
from blocksnet import LandUsePrediction

lup = LandUsePrediction(city_model=Villozskoe)
result_lup = lup.calculate()
LandUsePrediction.plot(result_lup)
```

Метод определения свободной площади в квартале

```python
from blocksnet import VacantArea
va = VacantArea(city_model=Villozskoe)
gdf_blocks = Villozskoe.get_blocks_gdf(simplify=True)
gdf_blocks.explore()
```

```python
block = Villozskoe[24]
result_va = va.calculate([block])
result_va.explore()
```
